# **<font color='#95f1c'>4. Train_Test_Split**

```python
# import the train_test_split function:

from sklearn.model_selection import train_test_split as TTS # Some people would not rename it but I did it to make it short.
```

```python
# Separate the Targe Value column (What we will predict).
y = df["SalePrice"]


# Now Separate the Features (The data our computer will learn in order to predict)
X = df.drop(columns = "SalePrice")
```

- we have saved our target in the variable "y".
- we have saved our features excluding our target in the var "X"

```python
# Now we do the split:
X_train, X_test, y_train, y_test = TTS(X,y, random_state = 42) # random_state = 42 its kind of the most used number to split
                                                               ## but there is also a formula to find the best random_state.
```

- Note how we will get 4 variables from the 2 "X" & "y" we created before.

# <font color='#95f1c'>**5. Imputing values**
- **What happens Before?**
- **What is the process?**

```python
# Select Numeric Columns
num_cols = X_train.select_dtypes("number").columns
num_cols
```

- **.select_dtypes("number").columns** will select the columns that hold numeric values.

### **Start the Imputer**

```python
from sklearn.impute import SimpleImputer

# start the inputer with the "median" strategy:
impute_median = SimpleImputer(strategy="median")
```

```python
# Fit the object on numeric training data with .fit():
impute_median.fit(X_train[num_cols])
```

```python
# check how many Null/NaNs are in the X_train numeric columns:
null_count = X_train[num_cols].isna().sum()
null_count
```

if you want to see only the ones that have nulls do this:

```python
null_count[null_count > 0] # filtering columns that have more than 0 nulls/NaNs
```

Notice that fitting the inputer didn't work, It is because we need to use the **.transform()** to really apply the imputer strategy. So everytime you think about imputing thing about **"transforming by imputing"**

```python
# Inpute Correctly:
X_train_numeric_Imputed = impute_median.transform(X_train[num_cols])
X_train_numeric_Imputed
```

**Is that what we wanted?**
- <font color='Red'> No, it is an array, and we need a Dataframe.

```python
# Change settings to ouput a dataframe
from sklearn import set_config
set_config(transform_output='pandas')
```

```python
# Transform the training data (fill in nulls with median)
X_train_num_imputed = impute_median.transform(X_train[num_cols])
X_train_num_imputed.head()
```

```python
X_train_num_imputed.isna().sum()
```

## Imputing Categorical and Ordinal columns/features


```python
# Define the "Categorical features" variable
cat_fish = X_train.select_dtypes("object").columns
cat_fish
```

```python
# Start the imputer:
cat_imputer = SimpleImputer(strategy = "constant", fill_value = "NA")
```

```python
# Fit the imputer on Training Data ONLY!!
cat_imputer.fit(X_train[cat_fish])
```

- **Tools you could use:**

```python
#Check what stragy U used:
print("The Stragy U used is:",cat_imputer.strategy.upper())    # .upper() just make any string output appear in UpperCase letters
```

```python
#Check What values you are going to impute:
print("The Stragy U used is:",cat_imputer.fill_value.lower())
```

```python
#Check what columns are you going to impute:
print("Columns you \".fit()\" on the imputer:",cat_imputer.feature_names_in_)
```

```python
# NaNs in Our Columns now?
X_train[cat_fish].isna().sum().sum()
```

```python
# Transform the NaNs/Nulls and save it in a new variable:
X_train_imputed = cat_imputer.transform(X_train[cat_fish])
```

```python
# Look for Nulls on the new X_train variable you created
X_train_imputed.isna().sum().sum()
```

### Now there are 0 Nulls/NaNs on the "numeric" and "categorical" data.
